### Problem Statement Scenario

<p1>Many social programs have a hard time ensuring that the right people are given enough aid. It’s tricky when a program focuses on the poorest segment of the population. This segment of the population can’t provide the necessary income and expense records to prove that they qualify.

In Latin America, a popular method called Proxy Means Test (PMT) uses an algorithm to verify income qualification. With PMT, agencies use a model that considers a family’s observable household attributes like the material of their walls and ceiling or the assets found in their homes to
classify them and predict their level of need.

While this is an improvement, accuracy remains a problem as the region’s population grows and poverty declines.

The Inter-American Development Bank (IDB) believes that new methods beyond traditional econometrics, based on a dataset of Costa Rican household characteristics, might help improve PMT’s performance. </p1>

```python
# Preprocessing of dependency column
data = train.dependency

plt.hist(data[data.isin(['no','yes'])==False].astype('float'),bins=50)
plt.show()

data.value_counts().head(10)
```

On comparing the frequency of values above in the histogram we can notice that the values starts from 0.5 has the highest frequency, which signifies that the mode is **0.5** .We can use mode value for imputation of data for categorical feature **dependency** 

For **`no=0`** & **`yes=0.5`**, can be imputed.

```python
train.dependency = train.dependency.replace('no',0).replace('yes',0.5).astype('float')
test.dependency = test.dependency.replace('no',0).replace('yes',0.5).astype('float')
```

```python
# Preprocessing of 'edjefe' column
data = train.edjefe

plt.hist(data[data.isin(['no','yes'])==False].astype('float'),bins=50)
plt.show()

data.value_counts().head(10)
```

On comparing the frequency values above in the histogram, we can notice that mode value is somewhere between **`5.0`** & **`7.5`** where it is very high. From above we can conclude that for **`6.0`** we have highest occurences. We can use this value for imputation of data for categorical feature **`edjefe`**  

**`no=0`** & **`yes=6.0`**

```python
train.edjefe = train.edjefe.replace('no',0).replace('yes',6.0).astype('float')
test.edjefe = test.edjefe.replace('no',0).replace('yes',6.0).astype('float')
```

```python
# Preprocessing of 'edjefa' column
data = train.edjefa

plt.hist(data[data.isin(['no','yes'])==False].astype('float'),bins=50)
plt.show()

data.value_counts().head(10)
```

On comparing the frequency values above in the histogram, we can notice that mode value is somewhere between **`5.0`** & **`7.5`** where it is very high. From above we can conclude that for **`6.0`** we have highest occurences. We can use this value for imputation of data for categorical feature **`edjefa`**  

**`no=0`** & **`yes=6.0`**

```python
train.edjefa = train.edjefa.replace('no',0).replace('yes',6.0).astype('float')
test.edjefa = test.edjefa.replace('no',0).replace('yes',6.0).astype('float')
```

```python
#Imputation of data for meaneduc and SQBmeaned features
from sklearn.impute import SimpleImputer
SimpleImputer = SimpleImputer(missing_values=np.nan, strategy='median')
SimpleImputer.fit(train[['meaneduc','SQBmeaned']])

train[['meaneduc','SQBmeaned']] = SimpleImputer.transform(train[['meaneduc','SQBmeaned']])
test[['meaneduc','SQBmeaned']] = SimpleImputer.transform(test[['meaneduc','SQBmeaned']])
```

### 9. Predict the accuracy using random forest classifier.

```python
y = train['Target']
X = train.drop(columns = ['Target'])
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

RFModel = RandomForestClassifier(min_samples_leaf=10,min_samples_split=10,random_state=1)
```

We will use **`GridSearchCV`** for hyperparameter tuning. We will tune:
    <li>max_depth</li>
    <li>min_samples_split</li>
    <li>min_samples_leaf</li>

```python
from sklearn.model_selection import GridSearchCV
# Create the parameter grid based on the results of random search 
param_grid = {
    'max_depth': [80, 90, 100, 110],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12]
}

# Instantiate the grid search model
grid_search = GridSearchCV(estimator = RFModel, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
```

```python
grid_search.fit(X,y)
```

```python
grid_search.best_params_
```

```python
best_grid = grid_search.best_estimator_
```

```python
# Predict on train data
pred_train = best_grid.predict(X)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
#Check the accuracy
print("\n\nAccuracy Score= ", accuracy_score(y_true=y,y_pred=pred_train),
      "\n\nConfusion Matrix \n", confusion_matrix(y_true=y,y_pred=pred_train),
      "\n\nClassification Report \n", classification_report(y_true=y,y_pred=pred_train)
     )
```

**Accuracy Score is: `92.88%`**

### 10. Check the accuracy using random forest with cross validation

```python
#Using cross validation to check accuracy:
from sklearn.model_selection import cross_val_score
cv = cross_val_score(best_grid,X,y,scoring='accuracy',cv=10)
print(np.mean(cv),"\t",min(cv),"\t",max(cv))
print("")
print(cv)
```

#### Maximum CV score is 0.67
